```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
import re

text_files = {
    'салаты': 'salads.txt',
    'рыбалка': 'fishing.txt',
    'походы': 'hiking.txt',
    'ремонт': 'repair.txt',
    'атмосферные явления': 'cyclone.txt'
}

texts = {}
for theme, file_path in text_files.items():
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        texts[theme] = text


def process_text(text):
    
    tokens = word_tokenize(text.lower())
    
    stemmer = SnowballStemmer('russian')
    tokens = [stemmer.stem(word) for word in tokens]
    
    stop_words = set(stopwords.words('russian'))
    tokens = [word for word in tokens if word not in stop_words]
    
    return tokens

with open('C:\\hws\\texttt.txt', 'r', encoding='utf-8') as f:
    input_text = f.read()

processed_input_text = process_text(input_text)

max_similarity = 0
selected_theme = ''
for theme, text in texts.items():
    processed_text = process_text(text)
    
    common_words = len(set(processed_input_text) & set(processed_text))
    print(common_words)
    
    if common_words > max_similarity:
        max_similarity = common_words
        selected_theme = theme
print('Вероятная тема текста:', selected_theme)
```

```python
import pymorphy2

morph = pymorphy2.MorphAnalyzer()

with open('textie.txt', 'r', encoding='UTF-8') as f:
    text = f.read()
words = text.split() 

for word in words:
    analyzed_word = morph.parse(word)[0]
    print(analyzed_word)
    
prepositions = ['в', 'во', 'за', 'меж', 'между', 'на', 'о', 'об', 'обо', 'по', 'под', 'подо', 'с', 'со']
prep_cases_count = {}

for i in range(len(words) - 1):
    if words[i].lower() in prepositions:  
        preposition = words[i].lower() 
        word_after_prep = morph.parse(words[i + 1])[0]  
        if word_after_prep.tag.case is not None: 
            case = str(word_after_prep.tag.case) 
            if preposition in prep_cases_count:
                if case in prep_cases_count[preposition]:
                    prep_cases_count[preposition][case] += 1  
                else:
                    prep_cases_count[preposition][case] = 1  
            else:
                prep_cases_count[preposition] = {case: 1}  


for preposition, cases_count in prep_cases_count.items():
    print(preposition, ':')
    for case, count in cases_count.items():
        print(case, ':', count)
```

```python
import pyconll

file = pyconll.load_from_file('GramEval2020-RuEval2017-social-dev.conllu')

for sentence in file:
    for token in sentence:
        if token.deprel == 'discourse':
            words = [token.form for token in sentence]
            print(" ".join(words))
            print(token.form, '(discourse)')
            discourse_found = True
        if token.deprel == 'dislocated':
            words = [token.form for token in sentence]
            print(" ".join(words))
            print(token.form, '(dislocated)')
            dislocated_found = True
if discourse_found == False:
    print('Не найден тег discourse')
if dislocated_found == False:
    print('Не найден тег dislocated')
            
        
            
        
```
