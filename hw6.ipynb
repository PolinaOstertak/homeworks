Задача 3.
Напишите игрушку "Тамагочи". У вас должно быть минимум три класса: два для питомцев, один для самой игры. При начале игры игрок должен выбирать, какое животное он хочет завести, и давать ему имя. Возраст и другие характеристики животного можно генерировать случайным образом. Животное можно кормить и играть с ним, еще оно спит. Сама игра должна проходить в цикле, на каждой итерации можно проверять, хочет ли животное спать, и отправлять его, все потребности можно реализовать через динамические атрибуты. Если животное долго не кормить, оно умирает.

```python
import random

class Cat:
    def __init__(self, name):
        self.name = name
        self.pet_type = 'Cat'
        self.age = random.randint(1, 5)
        self.hungry = random.randint(1, 5)
        self.happy = random.randint(0, 5)
        self.energy = random.randint(0, 5)
        self.sleepy = random.randint(0, 5)
    def feed(self):
        self.hungry -= 2
        if self.hungry < 0:
            self.hungry = 0
        elif self.hungry > 5:
            self.hungry = 5
    def play(self):
        self.happy += 1
        self.energy -= 1
        if self.happy > 5:
            self.happy = 5
        if self.energy < 0:
            self.energy = 0
    def sleep(self):
        self.sleepy -= 2
        if self.sleepy < 0:
            self.sleepy = 0
        self.energy = 5
    def alive(self):
        return self.hungry < 5
class Dog:
    def __init__(self, name):
        self.name = name
        self.pet_type = 'Dog'
        self.age = random.randint(1, 5)
        self.hungry = random.randint(1, 5)
        self.happy = random.randint(0, 5)
        self.energy = random.randint(0, 5)
        self.sleepy = random.randint(0, 5)
    def feed(self):
        self.hungry -= 2
        if self.hungry < 0:
            self.hungry = 0
    def play(self):
        self.happy += 1
        self.energy -= 1
        if self.energy < 0:
            self.energy = 0
    def sleep(self):
        self.sleepy -= 2
        if self.sleepy < 0:
            self.sleepy = 0
        self.energy = 5
    def alive(self):
        return self.hungry < 5
class Game:
    def __init__(self):
        self.pet = None
    def play(self):
        pet_type = input('собака или кошка: ')
        pet_name = input('имя: ')
        if pet_type == 'собака':
            self.pet = Dog(pet_name)
        elif pet_type == 'кошка':
            self.pet = Cat(pet_name)
        while self.pet.alive():
            print('1. покормить')
            print('2. поиграть')
            print('3. уложить спать')
            c = int(input())
            if c == 1:
                self.pet.feed()
            elif c == 2:
                self.pet.play()
            elif c == 3:
                self.pet.sleep()
            self.pet.energy -= 1
            if self.pet.energy < 0:
                self.pet.energy = 0
            self.pet.hungry += 1 
            self.pet.sleepy += 1
            if self.pet.sleepy > 5:
              self.pet.sleepy = 5
            print(f'уровень голода - {self.pet.hungry}, уровень счастья - {self.pet.happy}, уровень энергии - {self.pet.energy}, желание спать - {self.pet.sleepy}')
        print('Вы убили животинку')
game = Game()
game.play()
```
